# **CLAHE IMAGE**

**CLAHE dataset setup**

```python
import cv2
import os
import shutil
import yaml

# Mengambil lokasi dataset original dari object roboflow
sumber_dataset = dataset.location 
target_dataset = '/kaggle/working/LPG-Dataset-CLAHE'

def apply_clahe_to_image(image_path, save_path):
    """Membaca gambar, menerapkan CLAHE pada channel Lightness (L), lalu menyimpan."""
    img = cv2.imread(image_path)
    if img is None: return
    
    # Ubah ke ruang warna LAB agar warna tidak rusak
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l_channel, a_channel, b_channel = cv2.split(lab)

    # Terapkan CLAHE dengan ClipLimit 2.0 dan Grid 8x8
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    cl = clahe.apply(l_channel)

    # Gabungkan kembali dan kembalikan ke format BGR
    merged_lab = cv2.merge((cl, a_channel, b_channel))
    final_img = cv2.cvtColor(merged_lab, cv2.COLOR_LAB2BGR)

    cv2.imwrite(save_path, final_img)

print("Memulai proses transformasi CLAHE... Mohon tunggu.")

# Looping ke folder train, valid, test
for split in ['train', 'valid', 'test']:
    img_dir_in = os.path.join(sumber_dataset, split, 'images')
    lbl_dir_in = os.path.join(sumber_dataset, split, 'labels')
    
    img_dir_out = os.path.join(target_dataset, split, 'images')
    lbl_dir_out = os.path.join(target_dataset, split, 'labels')
    
    # Buat direktori output
    os.makedirs(img_dir_out, exist_ok=True)
    os.makedirs(lbl_dir_out, exist_ok=True)
    
    # 1. Proses Gambar (Terapkan CLAHE)
    if os.path.exists(img_dir_in):
        for img_name in os.listdir(img_dir_in):
            in_path = os.path.join(img_dir_in, img_name)
            out_path = os.path.join(img_dir_out, img_name)
            apply_clahe_to_image(in_path, out_path)
            
    # 2. Proses Label (Copy paste murni tanpa ubah koordinat)
    if os.path.exists(lbl_dir_in):
        for lbl_name in os.listdir(lbl_dir_in):
            shutil.copy2(os.path.join(lbl_dir_in, lbl_name), os.path.join(lbl_dir_out, lbl_name))

# 3. Update file data.yaml agar mengarah ke folder baru
with open(os.path.join(sumber_dataset, 'data.yaml'), 'r') as file:
    yaml_data = yaml.safe_load(file)

# Sesuaikan path di yaml untuk folder CLAHE
yaml_data['path'] = target_dataset
# Hapus absolute path jika ada agar YOLO menggunakan relative path dengan aman
for key in ['train', 'val', 'test']:
    if key in yaml_data and yaml_data[key].startswith('/'):
        yaml_data[key] = yaml_data[key].split('/')[-2:] # ambil format folder/images
        yaml_data[key] = '/'.join(yaml_data[key])

with open(os.path.join(target_dataset, 'data.yaml'), 'w') as file:
    yaml.dump(yaml_data, file, default_flow_style=False)

print(f"Selesai! Dataset CLAHE siap digunakan di: {target_dataset}/data.yaml")
```
